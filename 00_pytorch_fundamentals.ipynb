### One of the  most common errors in deep learning : shape errors
1. The **inner dimmensions** must match
* `(3, 2) @ (3, 2)` won't work
* `(2, 3)@ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work
2. The resulting matrix  has the shape of **outer dimmensions**:
* `(2,3) @ (3,2)` ->> `(2, 2)`
* `(3,2) @ (2,3)` ->> `(3, 3)`

```python
torch.matmul(torch.rand(3, 2), torch.rand(2, 3))
```

```python
# Shape for matrix multiplication
tensor_A = torch.tensor([[1,  2],
                         [3,  4],
                         [5,  6]])

tensor_B = torch.tensor([[7,  10],
                         [8,  11],
                         [9,  12]])

# torch.mm(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issues, we can manupulate the shape of one of our tensors using a **transpose**

A **transpose** switches the axes or a dimension of a given tensor

```python
tensor_B.T, tensor_B.T.shape
```

```python
tensor_B, tensor_B.shape
```

```python
torch.mm(tensor_A, tensor_B.T).shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape } (same shape as above ) , tensor_B.T = {tensor_B.T}")
print(f"After multiplying {tensor_A}*{tensor_B.T.shape} < - inner dimmension must match")
print(f"Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(f"\nThe output is {output} and the shape is {output.shape}")
```

## Reshaping, stacking, squeezing and unsqueezing tensors:
* Reshaping  - reshapes the input tensor to a defined shape
* View - return a view of input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on the top of each other (vstack) or side by side (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - adds a `1` dimension to a target column
* Permute - Return a view of a input with dimension permuted (swaped) in a certain wat

```python
# Let's create a Tensor
x = torch.arange(1. , 10.)
x, x.shape
```

```python
x_reshaped = x.reshape(1, 9)
x_reshaped
```

```python
# Change the review
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
#  stack tensors on the top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# torch.squeeze - removes all 1 dimensions from a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous Shape: {x_reshaped.shape}")

# Remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew Tensor : {x_squeezed}")
print(f"\nShape of new Tensor: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at specific dim (dimensions)
print(f"Previous Target : {x_squeezed}")
print(f"Previous Shape: {x_squeezed.shape}")
# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\n New Target : {x_unsqueezed}")
print(f"\n New Shape : {x_unsqueezed.shape}")
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at specific dim (dimensions)
print(f"Previous Target : {x_squeezed}")
print(f"Previous Shape: {x_squeezed.shape}")
# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=1)
print(f"\n New Target : {x_unsqueezed}")
print(f"\n New Shape : {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, color_channel]

# Permute to original  tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1 , 1->2, 2->0

print(f"Original Shape : {x_original.shape}")
print(f"\nPermuted Shape:{x_permuted.shape}") # [color_channel, height, width]
# Memory would be same
```

### Reproducbility (trying to take random out of random)
In short how a neural network learns:
`start with random numbers -> tensor operations -> update random numbers to try and make them better representation of the data -> again -> again -> again...`
To reduce the randomness in neural network and PyTorch comes the concept of a **random seed**

Especially what the random seeds does is "flavour" the randomness.

```python
torch.rand(3, 3)
```

```python
# Create two random tensor
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random tensors but reproducible tensors
import torch


# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)
torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C==random_tensor_D)
```
